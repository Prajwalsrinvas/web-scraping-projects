```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

location = input("Enter the place to search for:\n")
first_page_url = f"https://www.airbnb.co.in/s/{location}/homes"
df = pd.DataFrame(columns=['Stay Name', 'Stay Type', 'Stay Size','Amenities', 'Price', 'Rating', 'num of reviews', 'link'])
page_count = 0

def get_info_from_page(url):
    global page_count
    page = requests.get(url)
    soup = BeautifulSoup(page.text, parser='lxml')
    stays = soup.find_all('div', class_='_8ssblpx')

    for stay in stays:
        try:
            stay_name = stay.find('div', class_='_r6zroz').find('div', class_='_bzh5lkq').text.strip()
            stay_type = stay.find('div', class_='_r6zroz').find('div', class_='_167qordg').text.strip()
            stay_size = stay.find_all('div', class_='_kqh46o')[ 0].text.replace('·', '-').strip() # as '·' is wrongly displayed in the csv file 
            amenities = stay.find_all('div', class_='_kqh46o')[ 1].text.replace('·', '-')
            price = stay.find('span', class_='_1p7iugi').text.split(':')[1].replace('₹', 'Rs.')   # as '₹' is wrongly displayed in the csv file 
            rating = stay.find('span', class_='_10fy1f8').text
            num_of_reviews = stay.find('span', class_='_a7a5sx').text.strip().replace('(', '').replace(')', '') # as review count is written in ()
            stay_link = 'https://www.airbnb.co.in' + stay.find('a', class_='_gjfol0').get('href')

            index = len(df)
            df.loc[index] = [stay_name, stay_type, stay_size,amenities, price, rating, num_of_reviews, stay_link]
        except:
            pass  # as only places with all data available is valid,places with any missing data is skipped

    if soup.find('a', {'aria-label': 'Next'}): # if next button is present in the page
        next_page_url = 'https://www.airbnb.co.in' + soup.find('a', {'aria-label': 'Next'}).get('href') # store next page link
    else:
        next_page_url = None

        
    if next_page_url:
        page_count+=1
        print(f"Page {page_count} scraped")
        get_info_from_page(next_page_url)
    else:
        df.to_csv('airbnb_listings.csv', index=False)
        print(f"Page {page_count+1} scraped")
        print('All pages have been scraped and data is stored in csv file!')

get_info_from_page(first_page_url)
```
