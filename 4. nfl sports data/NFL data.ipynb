```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.nfl.com/standings/league/2019/reg/'
page = requests.get(url)
soup = BeautifulSoup(page.text,parser='lxml')
table = soup.find('table',class_="d3-o-table d3-o-table--row-striping d3-o-table--detailed d3-o-standings--detailed d3-o-table--sortable {sortlist: [[4,1]], sortinitialorder: 'desc'}")

# finding column names
thead = table.find('thead')
column_names = [column.text.strip() for column in thead.find_all('th')]
df = pd.DataFrame(columns = column_names)

rows = table.find('tbody').find_all('tr')

for row in rows:
    name = row.find('td').find('div',class_='d3-o-club-fullname').text.strip().replace('\n','')
    data = [d.text.strip() for d in row.find_all('td')[1:]] # as we have already stored names
    index = len(df)
    df.loc[index] = [name,*data]
    
df.to_csv('nfl_data.csv',index=False)
```
