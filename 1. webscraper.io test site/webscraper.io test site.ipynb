```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://webscraper.io/test-sites/e-commerce/allinone/phones/touch"
page = requests.get(url)
soup = BeautifulSoup(page.text,parser='lxml')

phones = soup.find_all('a',class_='title')
prices = soup.find_all('h4',class_='pull-right price')
descriptions = soup.find_all('p',class_='description')
ratings = soup.find_all('p',{'data-rating': True})
reviews = soup.find_all('p',class_='pull-right')

d = {'phones':[],
     'prices':[],
    'descriptions':[],
    'ratings':[],
    'reviews':[]}

for phone,price,desc,rating,review in zip(phones,prices,descriptions,ratings,reviews):
    d['phones'].append(phone.text)
    d['prices'].append(price.text)
    d['descriptions'].append(desc.text)
    d['ratings'].append(rating['data-rating'])
    d['reviews'].append(review.text.split()[0])
    
df = pd.DataFrame(d)
df
```
